```python
##Replacing null values using mean ,median or mode
```

```python
df = pd.read_csv('data_modified.csv')
df
```

```python
df.pH
```

```python
mean_pH = df.pH.mean()
mean_pH 
```

```python
mean_pH = df.pH.mean()
mean_density= df.density.mean()
mean_alcohol = df.alcohol.mean()
mean_chlorides = df.chlorides.mean()
```

```python
print(f'pH: {mean_pH}')
print(f'density: {mean_density}')
print(f'alcohol: {mean_alcohol}')
print(f'chlorides: {mean_chlorides}')
```

```python
df['pH'] =df['pH'].fillna(mean_pH)
df['density'] =df['density'].fillna(mean_density)
df['alcohol'] =df['alcohol'].fillna(mean_alcohol)
df['chlorides'] =df['chlorides'].fillna(mean_chlorides)
```

```python
df
```

```python
df.info()
```

```python
median_pH = df.pH.median()
median_density= df.density.median()
median_alcohol = df.alcohol.median()
median_chlorides = df.chlorides.median()
```

```python
print(f'pH: {median_pH}')
print(f'density: {median_density}')
print(f'alcohol: {median_alcohol}')
print(f'chlorides: {median_chlorides}')
```

```python
df['pH'] =df['pH'].fillna(median_pH)
df['density'] =df['density'].fillna(median_density)
df['alcohol'] =df['alcohol'].fillna(median_alcohol)
df['chlorides'] =df['chlorides'].fillna(median_chlorides)
```
